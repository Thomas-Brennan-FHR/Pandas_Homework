```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["Student ID"].count()
total_budget=school_data_complete["budget"].unique().sum()
avg_math=school_data_complete["math_score"].mean()
avg_reading=school_data_complete["reading_score"].mean()
overall_passing_rate=(avg_math+avg_reading)/2
passing_math_pct=(len([x for x in school_data_complete["math_score"] if x>=70])/total_students)*100
passing_reading_pct=(len([x for x in school_data_complete["reading_score"] if x>=70])/total_students)*100
overall_passing_rate=(passing_math_pct+passing_reading_pct)/2
district_summary=pd.DataFrame({
                                "Total Schools":[total_schools],
                                "Total Students":[total_students],
                                "Total Budget":[total_budget],
                                "Average Math Score":[avg_math],
                                "Average Reading Score":[avg_reading],
                                "% Passing Math":[passing_math_pct],
                                "% Passing Reading":[passing_reading_pct],
                                "% Overall Passing Rate":[overall_passing_rate]})
format_mapping={'Total Students': '{:,.0f}','Total Budget': '${:,.0f}', 'Average Math Score': '{:,.2f}', 'Average Reading Score': '{:,.2f}',
                '% Passing Math': '{:,.2f} %', '% Passing Reading': '{:,.2f} %', '% Overall Passing Rate': '{:,.2f} %'}
for key, value in format_mapping.items():
    district_summary[key] = district_summary[key].apply(value.format)
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
Top_schools=school_data_complete
f={"Student ID":["count"],"budget":["mean",lambda val: val.mean()/val.count()],"reading_score":["mean",lambda val: (val >= 70).sum()/val.count()*100],
   "math_score":["mean",lambda val: (val >= 70).sum()/val.count()*100]}
Top_schools=Top_schools.groupby(["school_name","type"]).agg(f)
Top_schools.columns = ['Total Students', 'Total School Budget','Per Student Budget','Average Reading Score','% Passing Reading','Average Math Score','% Passing Math']
Top_schools["% Overall Passing Rate"]=(Top_schools['% Passing Reading']+Top_schools['% Passing Math'])/2
Unformated_Schools=Top_schools*1
format_mapping={'Total Students': '{:,.0f}', 'Total School Budget': '${:,.0f}', 'Per Student Budget': '${:.2f}','Average Reading Score':'{:.2f}','% Passing Reading':'{:.2f} %',
                'Average Math Score':'{:.2f}','% Passing Math':'{:.2f} %','% Overall Passing Rate':'{:.2f} %'}
for key, value in format_mapping.items():
   Top_schools[key] = Top_schools[key].apply(value.format)

Top_schools=Top_schools.sort_values("% Overall Passing Rate",ascending=False)
Top_schools.head()
```
